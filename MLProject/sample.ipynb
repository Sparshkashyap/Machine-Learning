```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score,mean_absolute_error,mean_squared_error
import pandas as pd
import numpy as np
import  matplotlib.pyplot as plt 


# Load Dataset
df = pd.read_csv(r"C:\Users\Dell\Desktop\python\Machine Learning\Foundation\ML project\Students Performance Dataset.csv")
print(df.head())

# Input / Output
x = df[["Study_Hours_per_Week"]]
y = df["Total_Score"]

# Create a model
model = LinearRegression()

# Train model
model.fit(x,y)

# Testing the trained model 
predict_scored = model.predict(x)

# Valid Regression metrics
mae = mean_absolute_error(y,predict_scored)
mse = mean_squared_error(y,predict_scored)
rmse = np.sqrt(mse)
r2 = r2_score(y,predict_scored)

# show outputs

print(f"Mean Absolute Error is :",round(mae,2))
print(f"Mean Squared Error is :",round(mse,2))
print(f" Root Mean Squared Error is :",round(rmse,2))
print(f"Model Accuracy",round(r2,3))


#Histrogram

plt.figure(figsize=(10,6))
plt.hist(df["Final_Score"], bins=30,color="skyblue",edgecolor="black")
plt.title("Distribution of Final Exam Score")
plt.xlabel("Final Exam score")
plt.ylabel("Number of Students")
plt.grid(True)
plt.show()

#Scatter + Regression line

plt.figure(figsize=(10,6))
plt.scatter(x,y,color="blue",label="Actual Score")
plt.plot(x,predict_scored,color="red",label="Predicted score (Regression line)")
plt.title("Model prediction vs Actual Score")
plt.xlabel("Study hours per weak")
plt.ylabel("final Output")
plt.grid(True)
plt.show()

# Optional

new_hour = 9
predic = model.predict([[new_hour]])
print(f"Prredict the Exam Score :{predic}")

```
